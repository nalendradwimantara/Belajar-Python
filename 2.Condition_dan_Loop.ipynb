## 1. if else

Pengambilan keputusan **(if else)** tidak hanya digunakan untuk menentukan tindakan apa yang akan diambil sesuai dengan kondisi, tetapi juga digunakan untuk menentukan tindakan apa yang akan diambil/dijalankan jika kondisi tidak sesuai.

```python
nilai = 80
```

```python
if(nilai > 76):
    print("Selamat Anda Lulus")
else:
    print("Belajar Lagi Yaa Nak")
```


## 2. if elif
Pengambilan keputusan **(if elif)** merupakan lanjutan/percabangan logika dari kondisi *if*. Dengan **elif** kita bisa membuat kode program yang akan menyeleksi beberapa kemungkinan yang bisa terjadi. Hampir sama dengan kondisi **else**, bedanya kondisi *elif* **bisa banyak dan tidak hanya satu**.

```python
harga = 21000
```

```python
if(harga < 20000):
    print("Murah Banget")
elif(harga == 20000):
    print("Deal!")
elif(harga > 20000):
    print("Kemahalan Boss")
```

## 4. while
Pengulangan *While* Loop di dalam bahasa pemrograman Python dieksesusi statement berkali-kali selama kondisi bernilai benar atau **True**.

```python
langkah = 1
```

```python
while (langkah <= 8):
    print ("langkah ke: ", langkah)
    langkah = langkah + 1
```

```python
tahap = 1
while tahap <= 4:
    print (f"tahap ke: {tahap}")
    tahap += 1
```

```python
i = 0
while True:  # loop tanpa henti
    i += 1
    if i == 3:
        continue  # lewati angka 3
    if i > 5:
        break     # keluar dari loop saat i > 5
    print(i)
```

###### Fungsi Kuadrat

diberi $x$, dengan $y = x^2 + 2x - 15$

```python
x = 0
while (x <= 5):
    y = (x**2)+(2*x)-15
    print ("x =",x, " y =",y)
    x = x + 1
```


###### Fungsi Rekursif

diberi $X_0$, dengan $X_i$$_+$$_1$$ = 2(X_i) + 1$

```python
x0 = 0
mulai = 1

while (mulai <= 5):
    xp = (2*x0)+1
    print ("x",mulai, "akan bernilai:",xp)
    mulai = mulai + 1
    x0 = xp
```

## 5. Nested Loop
Bahasa pemrograman Python memungkinkan penggunaan satu *loop* di dalam *loop* lain. Pada materi kali ini diharapkan sudah dapat mulai **berfikir komputasional** dimana logika yang digunakan sesuai dengan logika berfikir komputer.

Mulailah dengan membangun **FlowChart** agar lebih memudahkan dalam berfikir komputasional seperti pada Program berikut ini.

Program pencari bilangan PRIMA

```python
angka = 2
stop = 20
while angka <= stop:
    is_prima = True
    pembagi = 2
    while pembagi < angka:
        if angka % pembagi == 0:
            is_prima = False
            break
        pembagi += 1
    if is_prima:
        print(angka, "adalah bilangan prima")
    angka = angka + 1
```

#### Tugas 3

```python
angka = int(input("masukan interval terkecil : "))
stop = int(input("masukan interval terbesar : "))
ganjil = []
genap = []
while angka <= stop:
    pembagi = 2
    if angka % pembagi == 0:
        genap.append(angka)
    else: 
        ganjil.append(angka)     
    angka = angka +1
    
print(genap, "adalah bilangan genap")
print(ganjil, "adalah bilangan ganjil")
```
