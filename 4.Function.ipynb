## 3. Fungsi dan Loop

diberi $x$, dengan $y = x^2 + 2x - 15$. tentukan $y$ pada interval $x[0,5]$

###### function dan for

```python
def kuadratik2(xawal, xakhir):
    for i in range(xawal, xakhir):
        y = (xawal**2)+(2*xawal)-15
        print ("x =",xawal, " y =",y)
        xawal = xawal + 1
    return
```

```python
kuadratik2(0,5)
```

###### function dan while

```python
def kuadratik3(xawal, xakhir):
    while (xawal <= xakhir):
        y = (xawal**2)+(2*xawal)-15
        print ("x =",xawal, " y =",y)
        xawal = xawal + 1
    return
```

```python
kuadratik3(xawal = 0, xakhir = 5)
```

### Bonus

Fungsi Bilangan Prima

```python
def Bilangan_Prima(angka, stop):
    while angka <= stop:
        is_prima = True
        pembagi = 2
        while pembagi < angka:
            if angka % pembagi == 0:
                is_prima = False
                break
            pembagi += 1
        if is_prima:
            print(angka, "adalah bilangan prima")
        angka = angka + 1
    return
```

```python
Bilangan_Prima(2,20)
```

##### Phytagoras

Segitiga Siku-siku dengan a = depan, b = samping, c = miring

```python
def pythagoras(a = 0, b = 0, c = 0):
    if (a == 0):
        a = sqrt((c**2) - (b**2))
        print("a =", a)
    elif (b == 0):
        b = sqrt((c**2) - (a**2))
        print("b =",b)
    elif (c == 0):
        c = sqrt((a**2) + (b**2))
        print("c =",c)
    return
```

```python
pythagoras(c = 10, b = 6)
```

###### Fungsi rekursif

diberi $X_0$, dengan $X_i$$_+$$_1$$ = 2(X_i) + 1$

```python
def rekursif(x0, i_awal, i_akhir):
    while (i_awal <= i_akhir):
        xp = (2*x0)+1
        print ("x",i_awal, " akan bernilai:",xp)
        i_awal = i_awal + 1
        x0 = xp
    return
```
