## 1. Membuat Matriks

Matriks adalah kumpulan bilangan yang tersusun dalam baris dan kolom. Matriks memiliki banyak kegunaan, salah satunya yakni memudahkan operasi sistem persamaan linear.

Isi matriks, disebut sebagai elemen, tersusun dalam 2-dimensi sehingga elemen matriks ditandai dengan dua indeks: letak baris lalu kolom. Dalam Python sendiri, matriks dapat dibuat dalam bentuk *list* dan *tuple*. Bedanya, elemen **tuple tidak dapat diubah** setelah dideklarasi dalam code sedangkan elemen **list masih dapat diubah.**

```python
A = [
     [2, 4, 6], 
     [4, 8, 9], 
     [0, 4, 3]
    ]
```

```python
A
```

Mengetahui jumlah baris dan kolom dengan perintah `len()`

```python
print("jumlah baris: ", len(A))
print("jumlah kolom: ", len(A[0]))
```

<div class="alert alert-block alert-danger">
    Mengetahui elemen pada matriks dengan <i>NESTED LOOP</i>
</div>



```python
for i in range(0, len(A)):
    for j in range(0, len(A[0])):
        print (A[i][j], end=' '),
    print
```

## 2. Penjumlahan dan Pengurangan Matriks

Sebagaimana bilangan, matriks memiliki operasi penjumlahan dan pengurangan. Dalam hal ini, ukuran tiap matriks dalam penjumlahan ataupun pengurangan harus sama. Operasi berlangsung antara elemen tiap matriks bersesuaian.

```python
B = [
     [1, -1, 3], 
     [-1, 1, 2], 
     [3, 2, 1]
    ]
```

```python
for i in range(0, len(A)):
    for j in range(0, len(A[0])):
        print (A[i][j] + B[i][j], end=' '),
    print
```

```python
for i in range(0, len(A)):
    for j in range(0, len(A[0])):
        print (A[i][j] - B[i][j], end=' '),
    print
```

```python
J = []

for i in range(0, len(A)):
    row = []
    for j in range(0, len(A[0])):
        jumlah = A[i][j] + B[i][j]
        row.append(jumlah)
    J.append(row)
```

```python
J
```

```python
K = []

for i in range(0, len(A)):
    row = []
    for j in range(0, len(A[0])):
        kurang = A[i][j] - B[i][j]
        row.append(kurang)
    K.append(row)
```

```python
K
```

## 3. Perkalian Matriks

Perkalian suatu matriks memang tidak sama dengan bilangan biasa, tidak juga langsung mengalikan setiap elemen. Perkalian matriks dilakukan dengan menjumlahkan hasil perkalian suatu baris matriks pertama ke kolom matriks kedua. Setiap baris di matriks pertama akan dikalikan ke setiap kolom di matriks kedua.

```python
M = []

for i in range(0, len(A)):
    row = []
    for j in range(0, len(A[0])):
        total = 0
        for k in range(0, len(A)):
            total = total + (A[i][k] * B[k][j])
        row.append(total)
    M.append(row)
```
